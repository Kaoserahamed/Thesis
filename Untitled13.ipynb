### Define Functions to Get Sentinel-1 and Sentinel-2 Collections

This cell defines two Python functions: `get_sentinel1_collection` and `get_sentinel2_collection`. These functions are used to filter and retrieve Sentinel-1 GRD and Sentinel-2 SR Harmonized image collections from Earth Engine based on a date range and the defined study area. The Sentinel-1 function includes a speckle filter, and the Sentinel-2 function selects and renames specific bands (Red, Green, Blue, NIR).

```python
# Function to get Sentinel-1 Collection
def get_sentinel1_collection(start_date, end_date, study_area):
    """
    Gets and filters the Sentinel-1 GRD image collection.

    Args:
        start_date (str): Start date for filtering (YYYY-MM-DD).
        end_date (str): End date for filtering (YYYY-MM-DD).
        study_area (ee.Geometry): The study area to filter the collection.

    Returns:
        ee.ImageCollection: Filtered Sentinel-1 image collection.
    """
    s1_collection = ee.ImageCollection('COPERNICUS/S1_GRD') \
        .filterDate(start_date, end_date) \
        .filterBounds(study_area) \
        .filter(ee.Filter.eq('instrumentMode', 'IW')) \
        .filter(ee.Filter.eq('orbitProperties_pass', 'DESCENDING')) \
        .filter(ee.Filter.eq('resolution_meters', 10)) \
        .filter(ee.Filter.eq('pol', 'VV'))

    # Apply speckle filter (focal_median)
    def apply_speckle_filter(image):
      return image.focal_median(50, 'square', 'meters')

    s1_collection = s1_collection.map(apply_speckle_filter)

    return s1_collection


# Function to get Sentinel-2 Collection
def get_sentinel2_collection(start_date, end_date, study_area):
    """
    Gets and filters the Sentinel-2 SR Harmonized image collection.

    Args:
        start_date (str): Start date for filtering (YYYY-MM-DD).
        end_date (str): End date for filtering (YYYY-MM-DD).
        study_area (ee.Geometry): The study area to filter the collection.

    Returns:
        ee.ImageCollection: Filtered Sentinel-2 image collection with renamed bands.
    """
    s2_collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
        .filterDate(start_date, end_date) \
        .filterBounds(study_area) \
        .filterMetadata('CLOUD_COVERAGE_ASSESSMENT', 'less_than', 10) \
        .select(['B4', 'B3', 'B2', 'B8'], ['Red', 'Green', 'Blue', 'NIR']) # Select and rename bands

    return s2_collection

print("Sentinel-1 and Sentinel-2 collection functions defined and updated for NDWI and band renaming.")
```

### Define Function to Get Representative Image Collection for a Year

This cell defines the `get_representative_image` function. This function attempts to retrieve a Sentinel-2 image collection for a given year and study area. If no Sentinel-2 images are found, it falls back to retrieving a Sentinel-1 image collection. It returns the collected image collection and a string indicating the sensor type.

```python
def get_representative_image(year, study_area):
    """
    Collects all available images (Sentinel-2 or Sentinel-1) for a given year.

    Prioritizes Sentinel-2, falls back to Sentinel-1.

    Args:
        year (int): The year for which to get the images.
        study_area (ee.Geometry): The study area.

    Returns:
        tuple: A tuple containing an ee.ImageCollection of available images
               and a string indicating the sensor ('Sentinel-2', 'Sentinel-1', or 'None').
    """
    start_date = f"{year}-01-01"
    end_date = f"{year}-12-31"

    # Try to get Sentinel-2 first (prioritize optical, less cloudy)
    s2_collection = get_sentinel2_collection(start_date, end_date, study_area)
    s2_size = s2_collection.size().getInfo()

    if s2_size > 0:
        print(f"Collected {s2_size} Sentinel-2 images for {year}.")
        return s2_collection, 'Sentinel-2'
    else:
        print(f"No suitable Sentinel-2 images found for {year}. Checking Sentinel-1.")
        # If no Sentinel-2, try Sentinel-1
        s1_collection = get_sentinel1_collection(start_date, end_date, study_area)
        s1_size = s1_collection.size().getInfo()

        if s1_size > 0:
            print(f"Collected {s1_size} Sentinel-1 images for {year}.")
            return s1_collection, 'Sentinel-1'
        else:
            print(f"No Sentinel-1 images found for {year} either.")
            # If neither is available
            return ee.ImageCollection([]), 'None' # Return an empty collection

print("get_representative_image function updated to collect all images.")
```

### Collect Representative Images for the Last Ten Years

This cell determines the last ten years and iterates through each year to collect representative image collections using the `get_representative_image` function. It stores the collected image collections, their corresponding sensor types, and the count of images for each year in dictionaries. Finally, it prints a summary of the image counts per year.

```python
from datetime import datetime

# Determine the range of the last ten years
current_year = datetime.now().year
start_year = current_year - 10
end_year = current_year # Include the current year

# Create an empty dictionary to store the representative image and sensor for each year
yearly_representative_images = {}
yearly_image_counts = {} # Dictionary to store image counts

# Loop through each year in the determined range
for year in range(start_year, end_year + 1):
    print(f"Collecting images for year {year}...")
    # Call the get_representative_image function
    image_collection, sensor = get_representative_image(year, study_area)

    # Store the returned image collection and sensor type in the dictionary
    if image_collection is not None:
        yearly_representative_images[year] = {'image': image_collection, 'sensor': sensor}
        yearly_image_counts[year] = image_collection.size().getInfo() # Store the image count
        # Removed the print statement inside the loop to print summary at the end
    else:
        yearly_image_counts[year] = 0
        print(f"No representative image collection collected for year {year}.")

print("\n--- Image Collection Summary ---")
for year, count in yearly_image_counts.items():
    print(f"Year {year}: {count} images collected")
print("--------------------------------")

print("\nYearly representative images collected.")
```

### Visualize Yearly Representative Images

This cell visualizes a sample image from the collected representative image collection for each year on a geemap interactive map. It calculates the centroid of the study area to center the map and applies appropriate visualization parameters based on whether the image is from Sentinel-2 (RGB) or Sentinel-1 (VV).

```python
# Calculate the centroid of the study area to center the map
centroid = study_area.centroid().coordinates().getInfo()
map_center = [centroid[1], centroid[0]] # geemap uses [latitude, longitude]

# Create a new map centered over the study area
Map_yearly_images = geemap.Map(center=map_center, zoom=8) # Use the calculated center

# Iterate through the collected yearly images and add them to the map
for year, image_info in yearly_representative_images.items():
    image_collection = image_info['image']
    sensor = image_info['sensor']

    # Check if the image collection is valid and not empty
    if image_collection is not None and image_collection.size().getInfo() > 0:
        # Select a random image from the collection for visualization
        # Clip the image to the study area before adding it to the map
        image = ee.Image(image_collection.toList(image_collection.size()).get(0)).clip(study_area) # Get the first image and clip

        if sensor == 'Sentinel-2':
            # Define visualization parameters for Sentinel-2 (RGB bands) - adjusted for scaled data
            # Use the renamed bands for visualization
            vis_params = {"bands": ["Red", "Green", "Blue"], "min": 0, "max": 3000} # Adjust max for typical visual range

        elif sensor == 'Sentinel-1':
            # Define visualization parameters for Sentinel-1 (VV band) - adjusted for scaled data
            # Sentinel-1 GRD in EE is typically in power. A range like 0 to 0.2 might be suitable before toB().
            # Let's use a slightly wider range based on the inspection to see if anything appears.
            vis_params = {"bands": ["VV"], "min": 0, "max": 0.2} # Adjusted example range for power scale

        else:
            # Skip if the sensor type is unexpected (shouldn't happen with current logic, but good practice)
            print(f"Skipping year {year}: Unknown sensor type '{sensor}'.")
            continue

        # Add the image to the map with a descriptive layer name
        Map_yearly_images.addLayer(image, vis_params, f"Year {year} ({sensor}) - Sample")
    else:
        print(f"Skipping year {year}: No valid image collection or collection is empty.")


# Display the map
Map_yearly_images
```

```python
# Visualize all available largest water body masks

import geemap
import ee

# Create a new map for visualizing all available water areas
# Center the map on the study area
centroid = study_area.centroid().coordinates().getInfo()
map_center = [centroid[1], centroid[0]]
Map_all_water = geemap.Map(center=map_center, zoom=8)


# Add a background layer for context (e.g., a recent Sentinel-2 image)
# Use the representative image for the latest year where one was collected
if yearly_representative_images: # Check if dictionary is not empty
    background_year = max(yearly_representative_images.keys())
    background_image_info = yearly_representative_images[background_year]

    if background_image_info['image'] is not None and background_image_info['sensor'] == 'Sentinel-2':
        # Compute median composite for background visualization
        background_image = background_image_info['image'].median().clip(study_area) # Clip background to study area
        # Define visualization parameters for Sentinel-2 (RGB bands)
        vis_params_background = {"bands": ["Red", "Green", "Blue"], "min": 0, "max": 3000}
        Map_all_water.addLayer(
            background_image,
            vis_params_background,
            f"Background Year {background_year} ({background_image_info['sensor']})"
        )
    elif background_image_info['image'] is not None and background_image_info['sensor'] == 'Sentinel-1':
         # Compute median composite for background visualization
        background_image = background_image_info['image'].median().clip(study_area) # Clip background to study area
        # Define visualization parameters for Sentinel-1 (VV band)
        vis_params_background = {"bands": ["VV"], "min": 0, "max": 0.2}
        Map_all_water.addLayer(
            background_image,
            vis_params_background,
            f"Background Year {background_year} ({background_image_info['sensor']})"
        )
else:
    print("No yearly representative images available to use as background.")

# Iterate through all years with collected representative images
for year, image_info in yearly_representative_images.items():
    image_collection = image_info['image']
    sensor = image_info['sensor']

    if image_collection is not None and image_collection.size().getInfo() > 0:
        if sensor == 'Sentinel-2':
            # Compute median composite for water mask calculation
            composite_image = image_collection.median()

            # Compute NDWI water mask
            s2_bands = composite_image.bandNames().getInfo()
            if 'Green' in s2_bands and 'NIR' in s2_bands:
                ndwi = composite_image.normalizedDifference(['Green', 'NIR']).rename('NDWI')
                water_mask = ndwi.gt(0).selfMask()
                # print(f"Year {year}: Initial water mask created (NDWI > 0).") # Removed for cleaner output during loop

                roi = study_area # Use the study_area variable as the ROI

                # Clip to ROI
                water_mask_clipped = water_mask.clip(roi)
                # print(f"Year {year}: Water mask clipped to ROI.") # Removed

                # Convert to vector polygons
                vectors = water_mask_clipped.reduceToVectors(
                    geometry=roi,
                    scale=30,
                    geometryType='polygon',
                    eightConnected=True,
                    maxPixels=1e8
                )
                # print(f"Year {year}: Water mask converted to vectors.") # Removed

                # Compute area for each polygon
                vectors = vectors.map(lambda f: f.set('area', f.geometry().area(ee.ErrorMargin(1))))
                # print(f"Year {year}: Area computed for each polygon.") # Removed

                # Sort by area and take the largest polygon
                largest_polygon = vectors.sort('area', False).first()

                # Check if a largest polygon was found
                if largest_polygon is not None:
                    print(f"Year {year}: Largest polygon found with area (m²):", largest_polygon.get('area').getInfo())

                    # Convert back to image mask for visualization
                    largest_polygon_image = ee.Image().paint(ee.FeatureCollection([largest_polygon]), 1).selfMask()
                    # print(f"Year {year}: Largest polygon converted back to image mask.") # Removed

                    # Add to map with a unique color for each year (optional, but helps distinguish)
                    # Using a simple palette for now, can be extended for more years
                    palette = {
                        2018: 'red',
                        2019: 'orange',
                        2020: 'yellow',
                        2021: 'green',
                        2022: 'blue',
                        2023: 'purple',
                        2024: 'cyan',
                        2025: 'magenta'
                        # Add more years and colors as needed
                    }
                    layer_color = palette.get(year, 'white') # Default to white if year not in palette

                    Map_all_water.addLayer(
                        largest_polygon_image,
                        {"palette": [layer_color]},
                        f"Largest Water Body {year}"
                    )
                    print(f"Year {year}: Largest Water Body mask added to map.")

                else:
                    print(f"Year {year}: No largest polygon found within the ROI. Skipping map layer.")

            elif sensor == 'Sentinel-1':
                print(f"Skipping water mask for year {year}: NDWI not directly applicable to Sentinel-1.")

        else:
            print(f"Skipping water mask for year {year}: No valid image collection or collection is empty.")

    else:
        print(f"Skipping water mask for year {year}: No representative image collection collected for this year.")

# Display the map
Map_all_water
```

### Compute and Visualize Largest Water Body Masks

This cell computes and visualizes the largest water body mask for specified years using the collected image collections. It prioritizes Sentinel-2 images for NDWI calculation to identify water. It then converts the water mask to vector polygons, calculates the area of each polygon, and selects the largest one. Finally, it converts the largest polygon back to an image mask and adds it to a new geemap interactive map, along with a background layer for context.

```python
# Re-run this cell to visualize the water masks with the corrected band names.
import geemap
import ee

# Years to visualize water area for
# Reduced to one year for faster processing/debugging
water_years = [2024]

# Create a new map for visualizing water area
Map_water = geemap.Map(center=[23.9, 90.4], zoom=8)

# Add a background layer for context (e.g., a recent Sentinel-2 image)
# Use the representative image for the latest year where one was collected
if yearly_representative_images: # Check if dictionary is not empty
    background_year = max(yearly_representative_images.keys())
    background_image_info = yearly_representative_images[background_year]

    if background_image_info['image'] is not None and background_image_info['sensor'] == 'Sentinel-2':
        # Compute median composite for background visualization
        background_image = background_image_info['image'].median()
        # Define visualization parameters for Sentinel-2 (RGB bands)
        # Use the renamed bands for background visualization
        vis_params_background = {"bands": ["Red", "Green", "Blue"], "min": 0, "max": 3000} # Adjust max for typical visual range
        Map_water.addLayer(
            background_image,
            vis_params_background,
            f"Background Year {background_year} ({background_image_info['sensor']})"
        )

    elif background_image_info['image'] is not None and background_image_info['sensor'] == 'Sentinel-1':
        # Compute median composite for background visualization
        background_image = background_image_info['image'].median()
        # Define visualization parameters for Sentinel-1 (VV band)
        vis_params_background = {"bands": ["VV"], "min": 0, "max": 0.2} # Adjust range based on data
        Map_water.addLayer(
            background_image,
            vis_params_background,
            f"Background Year {year} ({background_image_info['sensor']})" # Changed to current year for background consistency
        )
else:
    print("No yearly representative images available to use as background.")


# Iterate through the specified years and compute/visualize water mask
for year in water_years:
    if year in yearly_representative_images:
        image_info = yearly_representative_images[year]
        image_collection = image_info['image']
        sensor = image_info['sensor']

        if image_collection is not None and image_collection.size().getInfo() > 0:
            if sensor == 'Sentinel-2':
                # Compute median composite for water mask calculation
                composite_image = image_collection.median()

                # Compute NDWI water mask
                s2_bands = composite_image.bandNames().getInfo()
                if 'Green' in s2_bands and 'NIR' in s2_bands:
                    ndwi = composite_image.normalizedDifference(['Green', 'NIR']).rename('NDWI')
                    water_mask = ndwi.gt(0).selfMask()
                    print(f"Year {year}: Initial water mask created (NDWI > 0).")

                    # Use the study_area variable as the ROI
                    roi = study_area

                    # Clip to ROI
                    water_mask_clipped = water_mask.clip(roi)
                    print(f"Year {year}: Water mask clipped to ROI.")

                    # Convert to vector polygons
                    vectors = water_mask_clipped.reduceToVectors(
                        geometry=roi,
                        scale=30,  # Use 30 m to reduce load (S2 is 10 m, but 30 m is fine for large rivers)
                        geometryType='polygon',
                        eightConnected=True,
                        #labelProperty='water', # Removed labelProperty as it's not used in subsequent steps and can cause issues
                        maxPixels=1e8
                    )
                    print(f"Year {year}: Water mask converted to vectors.")

                    # Compute area for each polygon
                    vectors = vectors.map(lambda f: f.set('area', f.geometry().area(ee.ErrorMargin(1)))) # Added error margin
                    print(f"Year {year}: Area computed for each polygon.")

                    # Sort by area and take the largest polygon
                    largest_polygon = vectors.sort('area', False).first()

                    # Check if a largest polygon was found
                    if largest_polygon is not None:
                        print(f"Year {year}: Largest polygon found with area (m²):", largest_polygon.get('area').getInfo())

                        # Convert back to image mask for visualization - wrap the single feature in a FeatureCollection
                        largest_polygon_image = ee.Image().paint(ee.FeatureCollection([largest_polygon]), 1).selfMask()
                        print(f"Year {year}: Largest polygon converted back to image mask.")


                        # Add to map
                        Map_water.addLayer(
                            largest_polygon_image,
                            {"palette": ["blue"]}, # Using palette as before
                            f"Largest Water Body {year}"
                        )
                        print(f"Year {year}: Largest Water Body mask added to map.")

                    else:
                        print(f"Year {year}: No largest polygon found within the ROI. Skipping map layer.")


                else:
                    print(f"Skipping water mask for year {year}: Required Sentinel-2 bands (Green, NIR) not found in composite.")

            elif sensor == 'Sentinel-1':
                # Sentinel-1 is a SAR image, NDWI is not directly applicable
                print(f"Skipping water mask for year {year}: NDWI not directly applicable to Sentinel-1.")

        else:
            print(f"Skipping water mask for year {year}: No valid image collection or collection is empty.")

    else:
        print(f"Skipping water mask for year {year}: No representative image collection collected for this year.")

# Display the map
Map_water
```
